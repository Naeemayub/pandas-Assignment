```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
Data_Path = r'../assignment/states/'
all_CSV = glob.glob(Data_Path + "/*.csv")
```

```python
Data_list = []

for filename in all_CSV:
    df = pd.read_csv(filename, index_col=None, header=0)
    Data_list.append(df)

us_census = pd.concat(Data_list, axis=0, ignore_index=True)
```

```python
print(us_census.columns)
```

```python
print(us_census.dtypes)
```

```python
us_census.head(10)
```

```python
us_census.drop(columns= ['Unnamed: 0'], inplace=True )
```

```python
us_census.dtypes
```

```python
us_census[['Male', 'Female']] = us_census.GenderPop.str.split("_", expand=True)
```

```python
us_census.head(10)
```

```python
us_census.drop(columns = 'GenderPop', inplace=True)
```

```python
us_census.columns
```

```python
us_census['Income'] = us_census.Income.str.strip("$")
```

```python
us_census.head()
```

```python
us_census = us_census.replace("%", "", regex=True)
```

```python
us_census.head()
```

```python
us_census = us_census.replace(["F", "M"], "", regex=True)
```

```python
us_census.head()
```

```python
us_census.loc[:,"Hispanic":'Income'] = round(us_census.loc[:, "Hispanic": 'Income'].apply(pd.to_numeric),2)
```

```python
us_census.head()
```

```python
us_census.dtypes
```

```python
us_census["Male"] = us_census['Male'].astype(int)
```

```python
us_census.dtypes
```

```python
np.where(us_census.applymap(lambda x: x==''))
```

```python
us_census.drop(columns = "Female" , inplace=True)
```

```python
us_census["Female"] = us_census["TotalPop"]- us_census["Male"]
```

```python
us_census.head()
```

```python
us_census["Female"] = us_census['Female'].astype(int)
```

```python
us_census.dtypes
```

```python
import matplotlib.pyplot as plt
plt.scatter(us_census["Female"],us_census['Income'])
plt.xlabel('Salary $', fontsize=12)
plt.ylabel("Female_Population", fontsize=12)
plt.show()
```

```python
us_census.duplicated().any()
```

```python
us_census.drop_duplicates(inplace=True)
```

```python
us_census.duplicated().any()
```

```python
plt.scatter(us_census["Female"],us_census['Income'])
plt.xlabel('Salary $', fontsize=12)
plt.ylabel("Female_Population", fontsize=12)
plt.show()
```
